Импортируем все модули

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF
import scipy as sc
```

Считаем датафрейм как csv файл

```python
df = pd.read_csv('coinbase-extracted-september-to-october.csv')
```

Построим график наивысшей цены от времени, гистограмму и эмпирическую функцию распределения сначала для биткоина, а затем и для других основных криптовалют.

Графики для биткоина.

```python
dfBTC = df[df['Symbol'] == 'BTC']
ecdf = ECDF(dfBTC.High)
plt.step(ecdf.x, ecdf.y)
dfBTC.hist('High',bins = 100)
dfBTC.plot(x='Datetime',y=['High','Open','Low','Close'])
```
